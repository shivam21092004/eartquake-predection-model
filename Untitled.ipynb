```python
from sklearn.model_selection import train_test_split

# Select relevant columns
X = df[['Latitude(deg)', 'Longitude(deg)', 'Depth(km)', 'No_of_Stations']]
y = df['Magnitude(ergs)']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

# Train the linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

scores= {"Model name": ["Linear regression", "SVM", "Random Forest"], "mse": [], "R^2": []}

# Predict on the testing set
y_pred = regressor.predict(X_test)

# Compute R^2 and MSE
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

scores['mse'].append(mse)
scores['R^2'].append(r2)

print("R^2: {:.2f}, MSE: {:.2f}".format(r2, mse))
```

```python
# Predict on new data
new_data = [[33.89, -118.40, 16.17, 11], [37.77, -122.42, 8.05, 14]]
new_pred = regressor.predict(new_data)
print("New predictions:", new_pred)
```

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV file with space delimiter
df = pd.read_csv('Earthquake_Data.csv', delimiter=r'\s+')

new_column_names = ["Date(YYYY/MM/DD)",  "Time(UTC)", "Latitude(deg)", "Longitude(deg)", "Depth(km)", "Magnitude", 
                    "Magnitude_Category", "No_of_Stations", "Gap", "Close", "RMS", "SRC", "EventID"]

df.columns = new_column_names

# Convert magnitude column to categorical data
df['Magnitude_Category'] = pd.cut(df['Magnitude'], bins=[0, 5, 6, 7, np.inf], labels=['Minor', 'Moderate', 'Strong', 'Major'])

# Encode Magnitude Category
le = LabelEncoder()
df['Magnitude_Category_Encoded'] = le.fit_transform(df['Magnitude_Category'])

# Normalize latitude and longitude values
scaler = MinMaxScaler()
df[['Latitude(deg)', 'Longitude(deg)']] = scaler.fit_transform(df[['Latitude(deg)', 'Longitude(deg)']])

# Select features
X = df[['Latitude(deg)', 'Longitude(deg)', 'No_of_Stations']]
y = df['Magnitude_Category_Encoded']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Gaussian Naive Bayes model on the training data
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# Use the trained model to make predictions on the testing data
y_pred = gnb.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

# Calculate and print the confusion matrix and classification report
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', cm)

cr = classification_report(y_test, y_pred, labels=[0, 1, 2, 3], target_names=['Minor', 'Moderate', 'Strong', 'Major'])
print('Classification Report:\n', cr)
```

```python
# Create a scatter plot of actual vs predicted values
plt.figure(figsize=(8, 8))
plt.scatter(X_test['Longitude(deg)'], X_test['Latitude(deg)'], c=y_test, cmap='viridis')
plt.title('Actual Magnitude Category')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
print(" ")
plt.figure(figsize=(8, 8))
plt.scatter(X_test['Longitude(deg)'], X_test['Latitude(deg)'], c=y_pred, cmap='viridis')
plt.title('Predicted Magnitude Category')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
print(" ")


# Create a heatmap of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted Magnitude Category')
plt.ylabel('Actual Magnitude Category')
plt.show()
print(" ")

cr = classification_report(y_test, y_pred, labels=[0, 1, 2, 3], target_names=['Minor', 'Moderate', 'Strong', 'Major'], output_dict=True)
# Convert classification report dictionary to DataFrame
cr_df = pd.DataFrame(cr).transpose()

# Create bar plot of classification report scores
plt.figure(figsize=(8, 6))
sns.barplot(x=cr_df.index, y=cr_df['f1-score'])
plt.xlabel('Magnitude Category')
plt.ylabel('F1 Score')
plt.title('F1 Score by Magnitude Category')
plt.show()
print(" ")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Step 1: Load the dataset
df = pd.read_csv('earthquake_data.csv', sep=r'\s+', engine='python')

# Step 2: Clean column names and handle any extra spaces
df.columns = df.columns.str.strip()

# Rename columns for better readability
df.rename(columns={
    'Date(YYYY/MM/DD)': 'Date',
    'Time': 'Time(UTC)',
    'Latitude': 'Latitude(deg)',
    'Longitude': 'Longitude(deg)',
    'Depth': 'Depth(km)',
    'Mag': 'Magnitude',
    'Magt': 'Magnitude_Category',
    'Nst': 'No_of_Stations',
    'Gap': 'Gap',
    'Clo': 'Close',
    'RMS': 'RMS',
    'SRC': 'SRC',
    'EventID': 'EventID'
}, inplace=True)

# Step 3: Drop rows with missing values in 'Latitude' and 'Longitude'
df = df.dropna(subset=['Latitude(deg)', 'Longitude(deg)'])

# Step 4: Select features and target variables
# Using Depth, Magnitude, No_of_Stations, Gap, RMS as features
features = df[['Depth(km)', 'Magnitude', 'No_of_Stations', 'Gap', 'RMS']]

# The target variables are Latitude and Longitude
target_lat = df['Latitude(deg)']
target_lon = df['Longitude(deg)']

# Step 5: Split the data into training and testing sets
X_train, X_test, y_train_lat, y_test_lat, y_train_lon, y_test_lon = train_test_split(
    features, target_lat, target_lon, test_size=0.2, random_state=42
)

# Step 6: Train the Random Forest model for Latitude and Longitude prediction
lat_model = RandomForestRegressor(n_estimators=100, random_state=42)
lat_model.fit(X_train, y_train_lat)

lon_model = RandomForestRegressor(n_estimators=100, random_state=42)
lon_model.fit(X_train, y_train_lon)

# Step 7: Make predictions on the test set
y_pred_lat = lat_model.predict(X_test)
y_pred_lon = lon_model.predict(X_test)

# Step 8: Evaluate the model
mae_lat = mean_absolute_error(y_test_lat, y_pred_lat)
mse_lat = mean_squared_error(y_test_lat, y_pred_lat)
mae_lon = mean_absolute_error(y_test_lon, y_pred_lon)
mse_lon = mean_squared_error(y_test_lon, y_pred_lon)

print(f"Latitude MAE: {mae_lat}")
print(f"Latitude MSE: {mse_lat}")
print(f"Longitude MAE: {mae_lon}")
print(f"Longitude MSE: {mse_lon}")

# Step 9: Visualize predictions vs actual locations
plt.figure(figsize=(10, 6))

# Plot for Latitude
plt.subplot(1, 2, 1)
plt.scatter(y_test_lat, y_pred_lat, color='blue', label='Predicted vs Actual Latitude')
plt.xlabel('Actual Latitude')
plt.ylabel('Predicted Latitude')
plt.title('Latitude Prediction')
plt.legend()

# Plot for Longitude
plt.subplot(1, 2, 2)
plt.scatter(y_test_lon, y_pred_lon, color='green', label='Predicted vs Actual Longitude')
plt.xlabel('Actual Longitude')
plt.ylabel('Predicted Longitude')
plt.title('Longitude Prediction')
plt.legend()

plt.tight_layout()
plt.show()

# Step 10: Make predictions for new earthquake data
# Example new data: [Depth, Magnitude, No_of_Stations, Gap, RMS]
new_data = np.array([[10.0, 6.5, 15, 120, 0.8]])  # New earthquake data
predicted_lat = lat_model.predict(new_data)
predicted_lon = lon_model.predict(new_data)

print(f"Predicted Latitude: {predicted_lat[0]}")
print(f"Predicted Longitude: {predicted_lon[0]}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# File path to your earthquake data
file_path = r"C:\Users\shiva\OneDrive\Desktop\ml\Earthquake_data_processed.xlsx"

# Initialize variables for latitude, longitude, and affected area counts
longitudes = []
latitudes = []
category_counts = {}

# Try reading the file with a specific encoding and skip malformed lines
try:
    df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', on_bad_lines='skip')  # Use correct delimiter
except pd.errors.ParserError:
    print("Error parsing the CSV file. Check the delimiter and structure of the file.")

# Filter rows where 'latitude' and 'longitude' are within valid ranges
df = df[(df['Latitude(deg)'] >= -90) & (df['Latitude(deg)'] <= 90)]
df = df[(df['Longitude(deg)'] >= -180) & (df['Longitude(deg)'] <= 180)]

# Drop rows with missing 'latitude' or 'longitude'
df = df.dropna(subset=['Latitude(deg)', 'Longitude(deg)'])

# Convert 'latitude' and 'longitude' columns to numeric values
df['Longitude(deg)'] = pd.to_numeric(df['Longitude(deg)'], errors='coerce')
df['Latitude(deg)'] = pd.to_numeric(df['Latitude(deg)'], errors='coerce')

# Drop rows where 'longitude' or 'latitude' are NaN after conversion
df = df.dropna(subset=['Longitude(deg)', 'Latitude(deg)'])

# Add latitude and longitude to lists
longitudes = df['Longitude(deg)'].tolist()
latitudes = df['Latitude(deg)'].tolist()

# Update the category counts for affected areas if present
if 'affected_area' in df.columns:
    category_counts.update(df['affected_area'].value_counts().to_dict())

# Create a figure for the map
fig = plt.figure(figsize=(12, 10))

# Create a Basemap instance for the world map
m = Basemap(projection='mill', 
            llcrnrlat=-80, urcrnrlat=80, 
            llcrnrlon=-180, urcrnrlon=180, 
            lat_ts=20, resolution='c')

# Convert collected data into map coordinates
x, y = m(longitudes, latitudes)

# Scatter plot on the map
m.scatter(x, y, color='blue', marker='o', s=10, label='Affected Areas')

# Draw coastlines, countries, and continents
m.drawcoastlines()
m.fillcontinents(color='coral', lake_color='white')
m.drawmapboundary(fill_color='white')
m.drawcountries()

# Set plot title and show the legend
plt.title("All Affected Areas")
plt.legend(loc='upper left')
plt.show()

# If there are category counts (affected areas), plot a bar chart
if category_counts:
    plt.figure(figsize=(10, 6))
    plt.bar(category_counts.keys(), category_counts.values(), color='green')
    plt.title('Bar Graph of Affected Areas')
    plt.xlabel('Area')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```
