```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVR 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
data = pd.read_csv('preprocessed_dataset2.csv')
data['Date(YYYY/MM/DD)/Time'] = pd.to_datetime(data['Date(YYYY/MM/DD)/Time']).astype(np.int64) // 10**9
data_sample = data.sample(n=5000, random_state=42)  
X_sample = data_sample[['Latitude', 'Longitude', 'Depth', 'Date(YYYY/MM/DD)/Time']]
y_sample = data_sample['Mag']
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = LinearSVR(max_iter=10000, random_state=42)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f"Linear SVR R²: {r2:.2f}, MSE: {mse:.2f}")
random_lat = np.random.uniform(low=-90, high=90)  
random_lon = np.random.uniform(low=-180, high=180) 
random_depth = np.random.uniform(low=0, high=10)  
random_time_seconds = np.random.randint(0, 86400)
hours = random_time_seconds // 3600
minutes = (random_time_seconds % 3600) // 60
seconds = random_time_seconds % 60
formatted_time = f"{hours:02}:{minutes:02}:{seconds:02}"
def time_to_seconds(time_str):
    hours, minutes, seconds = map(int, time_str.split(':'))
    return hours * 3600 + minutes * 60 + seconds
new_data = pd.DataFrame([[random_lat, random_lon, random_depth, formatted_time]], 
                        columns=['Latitude', 'Longitude', 'Depth', 'Date(YYYY/MM/DD)/Time'])
new_data['Date(YYYY/MM/DD)/Time'] = new_data['Date(YYYY/MM/DD)/Time'].apply(time_to_seconds)
new_data_scaled = scaler.transform(new_data[['Latitude', 'Longitude', 'Depth', 'Date(YYYY/MM/DD)/Time']])
new_pred = model.predict(new_data_scaled)
print("\nNew random data:")
print(f"Latitude: {random_lat}")
print(f"Longitude: {random_lon}")
print(f"Depth: {random_depth}")
print(f"Time (in HH:MM:SS format): {formatted_time}")
print(f"Linear SVR prediction for the random data: {new_pred[0]}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.svm import SVC
style.use('fivethirtyeight')
df = pd.read_csv('preprocessed_dataset2.csv') 
if 'Mag' not in df.columns:
    print("Error: 'Mag' column not found!")
else:
    print("Column 'Mag' is found.")

if 'Latitude' not in df.columns:
    print("Error: 'Latitude' column not found!")
else:
    print("Column 'Latitude' is found.")
df['Magnitude_Category'] = pd.cut(df['Mag'], bins=[0, 4, 6, 8, 10], labels=['ml', 'mm', 'md', 'mwc'])
subset_size = min(500, len(df))
features = df[['Mag', 'Latitude']].values[:subset_size]
classes = df['Magnitude_Category'][:subset_size].values
print("Features shape:", features.shape)
print("Classes shape:", classes.shape)
def make_meshgrid(x, y, h = 0.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy
def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
svm1 = SVC(kernel='rbf')
svm2 = SVC(kernel='rbf')
svm3 = SVC(kernel='rbf')
svm4 = SVC(kernel='rbf')
if 'ml' in classes:
    svm1.fit(features, (classes == 'ml').astype(int))
if 'mm' in classes:
    svm2.fit(features, (classes == 'mm').astype(int))
if 'md' in classes:
    svm3.fit(features, (classes == 'md').astype(int))
if 'mwc' in classes:
    svm4.fit(features, (classes == 'mwc').astype(int))
fig, ax = plt.subplots()
X0, X1 = features[:, 0], features[:, 1]
xx, yy = make_meshgrid(X0, X1)
color = ['y', 'b', 'g', 'k', 'm']
for i in range(subset_size):
    if classes[i] == 'ml':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[0])
    elif classes[i] == 'mm':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[1])
    elif classes[i] == 'md':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[2])
    elif classes[i] == 'mwc':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[5])
    else:
        plt.scatter(features[i][0], features[i][1], s=20, c=color[4])
plt.title('SVM Classification of Earthquake Magnitude Types')
plt.xlabel('Mag (ergs)')
plt.ylabel('Latitude (deg)')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_recall_curve
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
try:
    df = pd.read_csv('preprocessed_dataset2.csv', sep=r'\s+')
except pd.errors.ParserError:
    df = pd.read_csv('preprocessed_dataset2.csv', delimiter=',')
df['Magnitude_Category'] = pd.cut(df['Mag'], bins=[0, 5, 6, 7, np.inf], labels=['Minor', 'Moderate', 'Strong', 'Major'])
le = LabelEncoder()
df['Magnitude_Category_Encoded'] = le.fit_transform(df['Magnitude_Category'])
scaler = MinMaxScaler()
df[['Latitude', 'Longitude']] = scaler.fit_transform(df[['Latitude', 'Longitude']])
df = df.drop(columns=['Magnitude_Category_Encoded'])
X = df[['Latitude', 'Longitude', 'Nst']]  
y = df['Magnitude_Category']  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rf = RandomForestClassifier(class_weight='balanced', random_state=42, n_estimators=100, max_depth=10)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
for class_label in le.classes_:
    if class_label in y_test.values:
        precision, recall, _ = precision_recall_curve(
            y_test == class_label, 
            rf.predict_proba(X_test)[:, le.transform([class_label])[0]]
        )        
        plt.figure(figsize=(8, 6))
        plt.plot(recall, precision, marker='.')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title(f'Precision-Recall Curve for {class_label} Class')
        plt.show()
```
